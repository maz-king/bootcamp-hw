### Observations

* Of the four drugs plotted, ketapril saw the largest increase in both tumor size and metastatic spread.

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = pd.read_csv("data/mouse_drug_data.csv")
clinical_trial_data_to_load = pd.read_csv("data/clinicaltrial_data.csv")

# Combine the data into a single dataset
combineddata = pd.merge(mouse_drug_data_to_load, clinical_trial_data_to_load, on="Mouse ID")

# Display the data table for preview

combineddata.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumormean = combineddata.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
tumordf = pd.DataFrame(tumormean)

# Preview DataFrame
tumordf.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumorsem = combineddata.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
tumorsemdf = pd.DataFrame(tumorsem)

# Preview DataFrame
tumorsemdf.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumorreset = tumordf.reset_index()
tumorreformat = tumorreset.pivot(index='Timepoint', columns='Drug')['Tumor Volume (mm3)']

tumorsemreset = tumorsemdf.reset_index()
tumorsemreformat = tumorsemreset.pivot(index='Timepoint', columns='Drug')['Tumor Volume (mm3)']

# Preview that Reformatting worked
tumorreformat
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(tumorreformat.index, tumorreformat["Capomulin"], yerr=tumorsemreformat["Capomulin"],\
             color="gold", marker="o", label="Capomulin")
plt.errorbar(tumorreformat.index, tumorreformat["Ceftamin"], yerr=tumorsemreformat["Ceftamin"],\
             color="lime", marker="s", label="Ceftamin")
plt.errorbar(tumorreformat.index, tumorreformat["Infubinol"], yerr=tumorsemreformat["Infubinol"],\
             color="navy", marker="d", label="Infubinol")
plt.errorbar(tumorreformat.index, tumorreformat["Ketapril"], yerr=tumorsemreformat["Ketapril"],\
             color="slategray", marker="^", label="Ketapril")

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)") 
plt.ylabel("Tumor Volume") 

plt.grid()
plt.legend(loc='best', fancybox=True)

# Save the Figure
plt.savefig("TumorResponse.png")

# Show the Figure
plt.show()
```
